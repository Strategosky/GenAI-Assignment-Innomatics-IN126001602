Task 1: Conceptual Understanding

**1. What is the difference between "Love" and "love" in NLP?**
Although they mean the same thing to a human, NLP systems treat "Love" and "love" as completely distinct tokens because their ASCII/Unicode values are different. If we don't convert text to lowercase, the model will create separate stochastic/vector representations for both, which wastes memory and dilutes the learned meaning of the word.

**2. What happens if stopwords are not removed?**
If common words like "the", "is", and "and" are not removed, they introduce noise into the dataset. Because they appear so frequently, they can overwhelm the meaningful words, reduce the accuracy of the model, and significantly increase the computational requirements needed to process the text.

**3. Provide two real-world scenarios where removing stopwords can be harmful.**

  Sentiment Analysis: Removing words like "not" or "very" completely changes the meaning of a sentence (e.g., "I am not happy" becomes "I am happy").

  Machine Translation / Conversational AI: To translate text accurately or generate human-like responses (like chatbots), the system needs grammatical context, which relies heavily on stopwords.

**4. Explain the difference between stemming and lemmatization.**

  Stemming is a blunt, rule-based approach that simply chops the ends off words to find the root. For example, it might chop "caring" into "car", which loses the actual meaning.

  Lemmatization uses a vocabulary and morphological analysis (a dictionary) to return a word to its proper base dictionary form (called a lemma). For example, it correctly changes "better" to "good", and "caring" to "care".

```python
#Task 2
import re

def preprocess_text(text):
    # Error Handling for Task 7: Check if text is empty or not a string
    if not text or not isinstance(text, str):
        return []

    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Handle repeated characters (e.g., "soooo" -> "so", "goooood" -> "good")
    # This regex looks for characters repeated more than twice and reduces them to a single character.
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove special characters/punctuation (leaves only words and spaces)
    text = re.sub(r'[^\w\s]', '', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Tokenize and remove very short tokens (length <= 2), keeping exceptions
    exceptions = {"no", "not", "ok", "go", "is", "it", "me", "am", "at", "to"}
    tokens = text.split()
    cleaned_tokens = [word for word in tokens if len(word) > 2 or word in exceptions]

    return cleaned_tokens

#test
print(preprocess_text("WOW!!! This is soooo GREAT!!! Visit http://example.com 123"))
```

```python
#Task 3
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Process and display results
processed_results = []
for text in sample_inputs:
    tokens = preprocess_text(text)
    clean_sentence = " ".join(tokens)
    processed_results.append({"original": text, "tokens": tokens, "clean": clean_sentence})

    print(f"Original Text: {text}")
    print(f"Cleaned Tokens: {tokens}")
    print(f"Cleaned Sentence: {clean_sentence}")
    print("-" * 50)
```

```python
#Task 4
sentence_stats = []

for i, res in enumerate(processed_results, 1):
    tokens = res['tokens']
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    # Store stats for comparison later
    sentence_stats.append({
        "original": res['original'],
        "total": total_tokens,
        "unique": unique_tokens,
        "avg": avg_len,
        "noise_score": len(res['original']) - len(" ".join(tokens)) # Difference in length shows noise removed
    })

    print(f"Sentence {i}: {res['original']}")
    print(f"  - Total Tokens: {total_tokens}")
    print(f"  - Unique Tokens: {unique_tokens}")
    print(f"  - Average Token Length: {avg_len:.2f}")
    print("-" * 30)

# Analysis Questions
print("\n--- Analysis Questions ---")

# Most noise = sentence with the largest difference between original and cleaned length
most_noise = max(sentence_stats, key=lambda x: x['noise_score'])
# Most meaningful = sentence with the highest number of tokens retained
most_meaningful = max(sentence_stats, key=lambda x: x['total'])

print(f"1. Which sentence had the most noise?")
print(f"   Answer: '{most_noise['original']}' (Removed URLs/Emojis/Special Chars)")

print(f"\n2. Which sentence retained the most meaningful tokens after cleaning?")
print(f"   Answer: '{most_meaningful['original']}' (Retained {most_meaningful['total']} tokens)")
```

```python
#Task 5
from collections import Counter

# Flatten the list of all tokens from Task 3
all_tokens = []
for res in processed_results:
    all_tokens.extend(res['tokens'])

word_counts = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(word_counts.most_common(10))

print("\nTop 5 Least Frequent Words:")
# Getting the least common items by slicing the most_common list from the end
print(word_counts.most_common()[:-6:-1])
```

```python
#Task 6
def full_pipeline(text_list):
    """
    Processes a list of strings and returns a dictionary
    containing all tokens and cleaned sentences.
    """
    pipeline_tokens = []
    pipeline_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        pipeline_tokens.append(tokens)
        pipeline_sentences.append(" ".join(tokens))

    return {
        "tokens": pipeline_tokens,
        "clean_sentences": pipeline_sentences
    }

# Testing the full pipeline
pipeline_output = full_pipeline(sample_inputs)
print("Pipeline Output (First 2 sentences):")
print({
    "tokens": pipeline_output["tokens"][:2],
    "clean_sentences": pipeline_output["clean_sentences"][:2]
})
```

```python
#Task 7
# Testing the engine against problematic inputs
error_test_cases = [
    "",                # Empty string
    "😊😇😍",           # Only emojis
    "1234567890",      # Only numbers
    "   "              # Only whitespace
]

print("Error Handling Results:")
for case in error_test_cases:
    result = preprocess_text(case)
    print(f"Input: '{case}' -> Tokens: {result}")
```

"The preprocessing engine treats emojis, numbers, and punctuation as noise. By removing them, we ensure the model only focuses on the core vocabulary (words), which makes the statistical analysis in Tasks 4 and 5 more accurate for language patterns."
