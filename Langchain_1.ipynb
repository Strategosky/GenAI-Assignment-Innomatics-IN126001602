```python
# Task 1
# We replace the f-string approach with LangChain's PromptTemplate

task1_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Generating the prompt (No f-strings used for the actual prompt logic)
prompt_1_output = task1_template.format(topic="Quantum Computing")
print("Task 1 Output:\n", prompt_1_output)
```

```python
# Task 2
task2_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

test_cases =[
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("--- Task 2 Outputs ---")
for tc in test_cases:
    print(task2_template.format(**tc))
```

```python
# Task 3
variations = {
    "Teaching": PromptTemplate(input_variables=["topic"], template="Explain {topic} clearly step by step"),
    "Interview": PromptTemplate(input_variables=["topic"], template="Ask 3 questions about {topic}"),
    "Storytelling": PromptTemplate(input_variables=["topic"], template="Explain {topic} as a story")
}

topic_input = "Machine Learning"

print("--- Task 3 Outputs ---")
for style, template in variations.items():
    print(f"[{style}]: {template.format(topic=topic_input)}")
```

```python
# Task 4
# This separates system behavior from user input
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert {role}."),
    ("user", "Please talk about {topic}.")
])

roles = ["teacher", "interviewer", "motivator"]
chat_topic = "Neural Networks"

print("--- Task 4 Outputs ---")
for role in roles:
    messages = chat_template.format_messages(role=role, topic=chat_topic)
    print(f"Role: {role}")
    # Displaying the structured message format LangChain uses
    for msg in messages:
        print(f"  {msg.type.capitalize()}: {msg.content}")
    print()
```

```python
# Task 5
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Assign defaults if invalid
    val_audience = audience if audience in valid_audiences else "beginner"
    val_tone = tone if tone in valid_tones else "casual"

    return val_audience, val_tone

# Task 6
def generate_prompt(topic, audience, tone, style):
    v_audience, v_tone = validate_inputs(audience, tone)

    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style.",
        "interview": "Conduct an interview about {topic} for an {audience} in a {tone} tone.",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style."
    }

    selected_template = style_templates.get(style.lower(), style_templates["teaching"])

    app_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template
    )

    return app_template.format(topic=topic, audience=v_audience, tone=v_tone)

print("--- Task 5 & 6 Outputs ---")
# 1. Valid Input test
print("Valid:", generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))

# 2. Invalid Input test (Fallback to 'beginner' and 'casual')
print("Invalid Fallback:", generate_prompt("Python", "alien", "angry", "teaching"))
```

```python
# Task 7
reusable_template = PromptTemplate(
    input_variables=["action", "subject", "context"],
    template="You need to {action} the {subject} considering {context}."
)

reusability_tests =[
    {"action": "optimize", "subject": "database", "context": "high traffic"},
    {"action": "feed", "subject": "cat", "context": "it is very hungry"},
    {"action": "debug", "subject": "code", "context": "a production error"},
    {"action": "motivate", "subject": "team", "context": "a tight deadline"},
    {"action": "secure", "subject": "API", "context": "recent vulnerabilities"}
]

print("--- Task 7 Outputs ---")
for i, inputs in enumerate(reusability_tests, 1):
    print(f"Test {i}: {reusable_template.format(**inputs)}")
```
